```python
from pathlib import Path
from typing import cast

from dotenv import dotenv_values

from ss_utils_bamboohr import BambooHRClient

BAMBOOHR_API_KEY = cast(str, dotenv_values()["BAMBOOHR_API_KEY"])
BAMBOOHR_COMPANY_DOMAIN = cast(str, dotenv_values()["BAMBOOHR_COMPANY_DOMAIN"])
BAMBOOHR_SESSION_COOKIE = cast(str, dotenv_values()["BAMBOOHR_SESSION_COOKIE"])

data_dir = Path("data")
job_app_dir = data_dir / "job_app"
job_app_dir.mkdir(parents=True, exist_ok=True)

# pyrefly: ignore[invalid-syntax]
async with BambooHRClient(
    company_domain=BAMBOOHR_COMPANY_DOMAIN,
    api_key=BAMBOOHR_API_KEY,
    session_cookie=BAMBOOHR_SESSION_COOKIE,
) as client:
    stats = await client.fetch_candidates_pipeline(
        output_base=job_app_dir, job_ids=[140, 141], rating_min=1.0
    )
    print(f"Processed {stats['processed']} candidates across all jobs.")
```
